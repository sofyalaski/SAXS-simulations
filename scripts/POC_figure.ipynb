# 1D scattering curve for a $4096^3$ fast Fourier transformed rebinned MOF structure.

No calculations are made in this file, everything was already processed, this is just an example of how 1D curve can look like
The file is available upon request, as it's a big one 

```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
import pandas as pd
```

```python
MOF_file = '/home/slaskina/1DMerged4096.nxs'
```

```python
with h5py.File(MOF_file, "r") as f:
    l = list(f['entry']['result'])
    print(l)
```

```python
with h5py.File(MOF_file, "r") as f:
    plt.plot(f['entry']['result']['Q'][()], f['entry']['result']['I'][()], color = 'red')
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel(r'$q (nm)^{-1}$',fontsize=14)
    plt.ylabel(r'$I (m\,sr)^{-1}$',fontsize=14)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
```
